```python
import os
import sys
import pandas as pd

sys.path.insert(0, os.path.abspath("../src"))

from cleaning import (
    clean_monthly_rent,
    clean_square_footage,
    clean_bedroom_count,
    clean_bathroom_count,
    extract_coordinates,
    clean_dates
)

df = pd.read_csv(
    "../data/raw/Rent_Board_Housing_Inventory_20260624.csv"
)
```

```python
df["square_footage_mid"] = clean_square_footage(
    df["square_footage"]
)

df[
    ["square_footage", "square_footage_mid"]
].head()
```

```python
df["bedroom_num"] = clean_bedroom_count(
    df["bedroom_count"]
)

df[
    ["bedroom_count", "bedroom_num"]
].head()
```

```python
df["bathroom_num"] = clean_bathroom_count(
    df["bathroom_count"]
)

df[
    ["bathroom_count", "bathroom_num"]
].head()
```

```python
coords = extract_coordinates(df["point"])

df = pd.concat([df, coords], axis=1)
```

```python
df = clean_dates(df)
df.info()
```
